```python
from diffusers import StableDiffusionPipeline
import torch

model_id = "sd-legacy/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

prompt = "a photo of an astronaut riding a horse on mars"
image = pipe(prompt).images[0]

image.save("astronaut_rides_horse.png")
```

```python
from diffusers import StableDiffusionPipeline
import torch

model_id = "sd-legacy/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

seed = 45
generator = torch.Generator("cuda").manual_seed(seed)

prompt = "a well organized living room with bright walls, pink sofa and dining table"
negative_prompt = "chair, flower vase"

image = pipe(prompt=prompt, negative_prompt=negative_prompt, generator=generator).images[0]
image.save("enhanced_scenery.png")
```
